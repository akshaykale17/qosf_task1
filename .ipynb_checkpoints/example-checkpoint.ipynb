```python
import importlib
import quantum_alternate_bits as qab
from qiskit import Aer, execute
importlib.reload(qab)
```

```python
#Ex 1
arr = [1,2,5,7] #Array with integer values
m = 3 #number of bits
n = 2 #number of address bits(2^n=len(arr))
```

```python
binary_arr = qab.convert_decimal_to_binary(arr,m)
print(binary_arr)
qc = qab.quantum_circuit(binary_arr,n,m)
qc.draw(output='mpl')
```

```python
qc.measure(range(n), range(n))
qc = qc.reverse_bits()
backend = Aer.get_backend('aer_simulator')
job = execute(qc, backend=backend, shots=1000, seed_simulator=12345)
result = job.result()
count =result.get_counts()
print(count)
two_states = []
two_states.append(max(count, key=count.get))
count.pop(two_states[0])
two_states.append(max(count, key=count.get))

print(two_states)
```
